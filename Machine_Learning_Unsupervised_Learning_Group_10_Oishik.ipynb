# **General Descriptions of Data**

* Data Source: The dataset is sourced from a comprehensive records of import and export transactions, which includes various attributes related to each transaction.

* Data Type: The dataset is structured as a tabular format (CSV) containing both categorical (e.g., Country, Product, Shipping_Method) and numerical (e.g., Quantity, Value, Weight) data types.

* Data Structure: The dataset consists of multiple columns including Transaction_ID, Country, Product, Import_Export (indicating whether the transaction is an import or export), Quantity, Value, Date, Category, Port, Customs_Code, Weight, Shipping_Method, Supplier, Customer, Invoice_Number, and Payment_Terms.ms.

#**Data Description**

A dataset of Imports-Exports is taken from kaggle.com https://www.kaggle.com/datasets/chakilamvishwas/imports-exports-15000) with 15000 rows x 16 columns,contains data related to international trade transactions. Each row represents a single transaction with the following columns:

***Columns***

* Transaction_ID: A unique identifier for each transaction.

* Country: The country involved in the transaction, either as the origin or destination.

* Product: The type of product being traded.

* Import_Export: Indicates whether the transaction is an import or export.

* Quantity: The quantity of the product traded.

* Value: The total value of the transaction in dollars.

* Date: The date the transaction occurred, in the format DD-MM-YYYY.

* Category: The category of the traded product, such as Machinery, Clothing, Electronics, Furniture, or Toys.

* Port: The port where the transaction took place.

* Customs_Code: A unique code assigned by customs authorities for the transaction.

* Weight: The total weight of the traded goods in kilograms.

* Shipping_Method: The method of shipping used for the transaction, either Air or Sea.

* Supplier: The company supplying the traded goods.

* Customer: The customer receiving the traded goods.

* Invoice_Number: A unique identifier for the invoice related to the transaction.

* Payment_Terms: The payment terms agreed upon for the transaction, such as Cash on Delivery, Prepaid, Net 30, Net 60, etc.

The data covers transactions from various countries and includes both imports and exports across different product categories. The transactions span from 2019 to 2024, with the majority occurring in 2021 and 2022.g in 2021 and 2022.

```python
import pandas as pd
import numpy as np
from scipy.stats import sem, t

# Assuming ob28_sdf_noncat is your DataFrame

# Create an empty dictionary to store the results
composite_results = {}

# Loop through each column in the DataFrame
for column in ob28_sdf_noncat.columns:
    # Calculate measures
    data = ob28_sdf_noncat[column].dropna()  # Remove missing values for calculations

    # Coefficient of Variation
    cv = (data.std() / data.mean()) * 100

    # Confidence Interval (95% confidence level)
    confidence_level = 0.95
    degrees_of_freedom = len(data) - 1

    # Calculate standard error of the mean
    sem_val = sem(data)

    # Calculate confidence interval
    confidence_interval = t.interval(confidence_level,
                                     degrees_of_freedom,
                                     loc=data.mean(),
                                     scale=sem_val)

    # Store the results in the dictionary
    composite_results[column] = {
        'Coefficient of Variation': cv,
        'Confidence Interval (95%)': confidence_interval
    }

# Display the results
for column, stats in composite_results.items():
    print(f"Variable: {column}")
    for stat_name, stat_value in stats.items():
        print(f"  {stat_name}: {stat_value}")
    print("-" * 20)
```

The analysis of the non-categorical variables in the dataset provides key insights through the calculation of the Coefficient of Variation (CV) and the 95% Confidence Interval for each variable. The Coefficient of Variation, which expresses the standard deviation as a percentage of the mean, indicates the relative variability of each variable; a higher CV suggests greater dispersion relative to the mean. The 95% Confidence Interval offers a range within which we can be 95% confident that the true population mean lies, providing an essential measure of uncertainty associated with the sample mean. ### Non Categorical Inferential Statistics

```python
import pandas as pd
from scipy.stats import shapiro, kstest, normaltest, jarque_bera

# Remove categorical variables from the dataset
ob28_noncat_variables = ob28_sdf_noncat

# Define a significance level
alpha = 0.05

# Perform normality tests for each non-categorical variable
for column in ob28_noncat_variables.columns:
    data = ob28_noncat_variables[column].dropna()  # Drop missing values

    print(f"Normality Test Results for '{column}':")

    # Shapiro-Wilk Test
    shapiro_stat, shapiro_p = shapiro(data)
    print(f"  Shapiro-Wilk Test: Statistic={shapiro_stat:.4f}, P-value={shapiro_p:.4f}, Normal={shapiro_p > alpha}")

    # Kolmogorov-Smirnov Test
    ks_stat, ks_p = kstest(data, 'norm', args=(data.mean(), data.std()))
    print(f"  Kolmogorov-Smirnov Test: Statistic={ks_stat:.4f}, P-value={ks_p:.4f}, Normal={ks_p > alpha}")

    # Anderson-Darling Test
    ad_test = normaltest(data)
    ad_stat, ad_p = ad_test.statistic, ad_test.pvalue
    print(f"  Anderson-Darling Test: Statistic={ad_stat:.4f}, P-value={ad_p:.4f}, Normal={ad_p > alpha}")

    # Jarque-Bera Test
    jb_stat, jb_p = jarque_bera(data)
    print(f"  Jarque-Bera Test: Statistic={jb_stat:.4f}, P-value={jb_p:.4f}, Normal={jb_p > alpha}")

    print("-" * 50)
```

The normality tests conducted on the non-categorical variables—Quantity, Value, and Weight—indicate a clear deviation from normal distribution across all measures. The Shapiro-Wilk test yielded statistics of 0.9574, 0.9548, and 0.9520 for each variable, respectively, all with P-values of 0.0000, leading to the conclusion that these variables are not normally distributed. Similarly, the Kolmogorov-Smirnov test results showed significant deviations with respective statistics of 0.0569, 0.0600, and 0.0616, also resulting in P-values of 0.0000. The Anderson-Darling test further supported this finding with extremely high test statistics (3423.4620 for Quantity, 4080.0370 for Value, and 5231.8689 for Weight), all with P-values of 0.0000. Lastly, the Jarque-Bera test confirmed the non-normality with high statistics and P-values of 0.0000 for all three variables. Overall, these results consistently indicate that Quantity, Value, and Weight do not follow a normal distribution, which may have implications for subsequent statistical analyses that assume normality.
